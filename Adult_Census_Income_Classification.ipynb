```python
## implementing single hidden layer neural network of layer_dims = [14,5,1] ##
def sigmoid(x):
    p = 1/(1+np.exp(-x))
    return p
def relu(np_arr):
    z = np.zeros(np_arr.shape)
    return np.maximum(z,np_arr)
            
```

```python
layer_dims = [10,5,1]
```

```python
def param_init(layer_dims):
    n_x = layer_dims[0]
    n_h = layer_dims[1]
    n_y = layer_dims[2]
    W1 = np.random.randn(n_h,n_x)*0.01
    #W1 = np.zeros([n_h,n_x])
    b1 = np.zeros([n_h,1])
    W2 = np.random.randn(n_y,n_h)*0.01
    #W2 = np.zeros([n_y,n_h])
    b2 = np.zeros([n_y,1])
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    return parameters
```

```python
parameters = param_init(layer_dims)
parameters["b1"]
```

```python
def forward(X,parameters):
    W1 = parameters['W1']
    b1 = parameters['b1']
    W2 = parameters['W2']
    b2 = parameters['b2']
    
    Z1 = W1.dot(X)+b1
    A1 = relu(Z1)
    Z2 = W2.dot(A1)+b2
    A2 = sigmoid(Z2)
    cache = {"Z1": Z1,
             "A1": A1,
             "Z2": Z2,
             "A2": A2}
    return A2,cache
```

```python
A2,cache = forward(train_x,parameters)
cache['Z2'].shape
```

```python
def computeCost(A2,Y,parameters):
    m = Y.shape[0]
    ##cache = forward(X,parameters)
    ##A2 = cache['A2']
    logprobs = np.log(A2)*Y
    prob1 = (1-Y)*(np.log(1-A2));
    cost = -(np.sum(logprobs)+np.sum(prob1))/m
    return cost
```

```python
cost = computeCost(A2,train_y,parameters)
```

```python
cost
```

```python
def gradRelu(Z):
    #Z[Z>0]=1
    #Z[Z<=0]=0
    x = (Z>0)
    return x
```

```python
def backward(X,Y,cache,parameters):
    m = X.shape[1]
    W1 = parameters['W1']
    W2 = parameters['W2']
    A1 = cache['A1']
    A2 = cache['A2']
    Z1 = cache['Z1']
    #Z11 = Z1
    
    dZ2 = A2-Y
    dW2 = dZ2.dot(A1.T)/m
    db2 = np.sum(dZ2,axis=1,keepdims=True)/m
    
    dZ1 = (W2.T).dot(dZ2)*gradRelu(Z1)
    dW1 = dZ1.dot(X.T)/m
    db1 = np.sum(dZ1,axis=1,keepdims=True)/m
    
    grads = {"dW1": dW1,
             "db1": db1,
             "dW2": dW2,
             "db2": db2}
    return grads
```

```python
grads = backward(train_x,train_y,cache,parameters)
grads
```

```python
def update(parameters,grads,alpha):
    W1 = parameters['W1']
    b1 = parameters['b1']
    W2 = parameters['W2']
    b2 = parameters['b2']
    
    dW1 = grads['dW1']
    db1 = grads['db1']
    dW2 = grads['dW2']
    db2 = grads['db2']
    
    W1 = W1 - alpha*dW1
    W2 = W2 - alpha*dW2
    b1 = b1 - alpha*db1
    b2 = b2 - alpha*db2
    
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters
```

```python
def nn_model(X,Y,alpha,nIter):
    m = X.shape[1]
    layer_dims=[10,8,1] 
    costs=[]
    parameters = param_init(layer_dims)
    for i in range(nIter):
        A2,cache = forward(X,parameters)
        costs.append(computeCost(A2,Y,parameters))
        grads = backward(X,Y,cache,parameters)
        parameters = update(parameters,grads,alpha)
    return parameters,costs
```

```python
parameters,costs = nn_model(train_x,train_y,alpha=.01,nIter=20000)
```

```python
def predict(X,parameters):
    m = X.shape[1]
    A2,cache = forward(X,parameters)
    Ypred = (A2[0]>.5)
    return Ypred.astype(int)
```

```python
def accuracy(Y,Yhat):
    A = Yhat-Y
    m = A.shape[0]
    count=0
    for i in range(0,m):
        if A[i] == 0:
            count = count+1
            
    return m,(count/m)*100
```
