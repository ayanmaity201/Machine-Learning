```python
def sigmoid(x):
    z = 1/(1+np.exp(-x))
    return z
```

```python
sigmoid(np.array([1,2,3,4]))
```

```python
def Initialize(X):
    n = X.shape[0]
    W = np.zeros([n,1])
    b = 0
    return W,b
```

```python
W,b = Initialize(X)
```

```python
def forward(X,W,b):
    z = (W.T).dot(X)+b
    Y = sigmoid(z.T)
    return Y
```

```python
Yhat = forward(X,W,b)
```

```python
Yhat
```

```python
(Yhat-Y)
```

```python
def computeCost(X,Yhat,Y):
    m = X.shape[1]
    J = -np.sum(Y*(np.log(Yhat))+(1-Y)*(np.log(1-Yhat)))/m
    return J
```

```python
cost = computeCost(X,Yhat,Y)
cost
```

```python
def derivative(X,Yhat,Y):
    m = X.shape[1]
    dW = X.dot(Yhat-Y)/m
    db = np.sum(Yhat-Y)/m
    return dW,db
```

```python
dW,db = derivative(X,Yhat,Y)
dW,db
```

```python
def update(W,b,dW,db,alpha):
    W = W - alpha*dW
    b = b - alpha*db
    return W,b
```

```python
w1,b1 = update(W,b,dW,db,.01)
w1,b1
```

```python
def gradient_descent(X,Y,W,b,nIter,alpha):
    cost=[]
    iter1 = []
    arrb = []
    for i in range(0,nIter):
        Yhat = forward(X,W,b)
        iter1.append(i)
        cost.append(computeCost(X,Yhat,Y))
        dW,db = derivative(X,Yhat,Y)
        W,b = update(W,b,dW,db,alpha)
        arrb.append(db)
    return W,b,cost,iter1,arrb
```

```python
W,b,cost,iter1,arrb = gradient_descent(X,Y,W,b,20000,0.003)
```

```python
def predict(X,W,b):
    m = X.shape[1]
    Yhat = forward(X,W,b)
    for i in range(0,m):
        if Yhat[i]>.5:
            Yhat[i] = 1
        else:
            Yhat[i] = 0
    return Yhat
```

```python
def accuracy(Y,Yhat):
    A = Yhat-Y
    m = A.shape[0]
    count=0
    for i in range(0,m):
        if A[i] == 0:
            count = count+1
            
    return (count/m)*100
```

```python
Yhat=predict(X,W,b)
```

```python
p = accuracy(Y,Yhat)
p
```
